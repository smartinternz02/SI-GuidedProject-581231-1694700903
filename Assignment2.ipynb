 Taking alcohol on x-axis and total on y-axis

# ***Distribution plot***

```python
sns.distplot(dataset["speeding"])
```

# ***Inference***
A distribution plot for "speeding" attribute is obtained
We can draw conclusions about the prevalence of speeding in car accidents based on the dataset, such as whether it follows a normal distribution, whether there are prominent peaks or clusters of speeding incidents, and whether there are any outliers or unusual patterns.

# ***Histogram***

```python
sns.histplot(data=dataset, x='alcohol', y='total')
```

# **Inference**
Taking "alcohol" attribute on x-axis and "total" attribute on y-axis
A histogram is plotted
Gaps or empty regions in the histogram, which could indicate missing or incomplete data.

# ***Relationplot***

```python
#using categorical attribute as hue
sns.relplot(x="alcohol",y="total",data=dataset,hue="abbrev")
```

# **Inference**
Taking alcohol on x-axis and total on y-axis
A relation plot is obtained
The relational plot (relplot) generates a scatter plot to depict the relationship between "alcohol" and "total" involvement in car accidents, with points coloured by the "abbrev" variable (likely representing different states or regions).Plot enables you to identify potential trends, differences, or outliers in the dataset by allowing you to understand the interplay between alcohol and speeding in car crashes across different states or regions.

# ***Barplot***

```python
sns.barplot(data=dataset,x="alcohol",y="total",ci=None)
```

# **Inference**
A barplot is obtained with "alcohol" on x-axis and "total" on y-axis
This bar plot compares the increasing levels of total involvement in car crashes with and without alcohol involvement, providing insight into the potential relationship between these two factors.

# ***CountPlot***

```python
sns.countplot(x="alcohol",data=dataset)
```

# **Inference**
A countplot is obtained for "alcohol" attribute
This graph depicts the distribution of car accidents involving and not involving consumption of alcohol. It enables you to estimate the frequency of influence of alcohol-related crashes.

# ***JOINTPLOT***

```python
sns.jointplot(x="alcohol",y="speeding",data=dataset)
```

# **Inference**
A jointplot is obtained by taking "alcohol" on x-axis and "speeding" on y-axis
The distribution of both variables is depicted in the plot, indicating the frequency of alcohol and speeding involvement. The scattered points and histograms aid in the identification of any patterns or correlations between the two variables, providing insight into their potential association in car accidents.

# ***BOXPLOT***

```python
sns.boxplot(x="alcohol",y="total",data=dataset)
```

# **Inference**
A boxplot is obtained with "alcohol" on x-axis and "total" on y-axis
This box plot compares the distribution, central tendency, and spread of total car accidents with and without alcohol involvement, assisting you in understanding potential differences or patterns between these two factors.

# ***CORELATION***
